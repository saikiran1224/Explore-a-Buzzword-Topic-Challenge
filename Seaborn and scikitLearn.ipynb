```python
# it contains lots of Data to practice
from sklearn.datasets import load_boston
```

```python
X,chacha = load_boston(return_X_y=True)
```

```python
# Violin plot shows the Probability density of the Data at Various values
sns.violinplot(y=chacha)
```

```python
# Outlier - something which is different from The whole Data 
# Consider the speed of Ussain Bolt when compared with other People 
# In a class only one Person has heoght of 7 ft others are approximately lie between 5 and 6 ft

#Outliers like "Wrong Data that does not belong here, They are Error Data"
# There are two types of Outliers
# 1. Fraud Data   - in reality the data is not clear 
# 2. Removed Data 
```

```python
# It directly shows the Outliers at the top and Bottom of the plot
# We simply remove the outliers for correct Data
sns.boxplot(y=chacha)
```
